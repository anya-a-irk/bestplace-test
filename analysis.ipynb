## Вопросы

#### Сколько магазинов формата стрит в тестовой выборке?

```python
test = ((df['Выборка'] == 'Тестовая') | (df['Выборка'] == 'Тест'))
street = ((df['Формат магазина'] == 'Street') | (df['Формат магазина'] == 'Стрит') | (df['Формат магазина'] == 'Strееt') | (df['Формат магазина'] == 'Стрт'))
print("Магазинов формата стрит в тестовой выборке", len(df[test & street]))
```

#### Какова средняя выручка магазинов формата мини ТЦ за 2016 год?

```python
all_data['2016'] = all_data[[x for x in all_data.columns if "2016" in x]].sum(axis=1)
```

```python
mini = (all_data['Формат магазина'] == 'Мини ТЦ') | (all_data['Формат магазина'] == 'Мини-ТЦ')
print("Средняя выручка магазинов формата мини ТЦ за 2016 год:", all_data[mini]['2016'].mean())
```

#### Сколько магазинов с бесплатной парковкой?

```python
free_parking = len(df[(df['Парковка'] == 'бесплатная парковка') | (df['Парковка'] == 'бесплатная паpковка')])
print("Магазинов с бесплатной парковкой", free_parking)
```
